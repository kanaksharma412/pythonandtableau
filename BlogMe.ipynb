```python
#dropping column
data = data.drop('engagement_comment_plugin_count', axis = 1)
```

```python
#keyword flagging

def KeyWordFlag(keyword):
    length = len(data)
    keyword_flag = []
    for x in range(0, length):
        heading = data['title'][x]
        try:
            if keyword in heading:
                is_present = True
            else:
                is_present = False
        except:
            is_present = False
        keyword_flag.append(is_present)
        
    return keyword_flag


KeyWordFlag = KeyWordFlag('murder')           
KeyWordFlag
```

```python
data['keyword_flag'] = pd.Series(KeyWordFlag) 
```

```python
# Sentiment Analysis using VADER

sent_int = SentimentIntensityAnalyzer()
positive = []
negative = []
neutral = []
length = len(data)
for i in range(0, length):
    try:
        text = data['title'][i]
        sent = sent_int.polarity_scores(text)
    except:
        positive.append(0)
        negative.append(0)
        neutral.append(0)
    positive.append(sent['pos'])
    negative.append(sent['neg'])
    neutral.append(sent['neu'])
    
```

```python
data['title_pos_sentiment'] = pd.Series(positive)
data['title_neg_sentiment'] = pd.Series(negative)
data['title_neu_sentiment'] = pd.Series(neutral)
```

```python
#writing the data
data.to_excel('blogme_clean.xlsx', sheet_name = 'blogmedata', index=False)
```
